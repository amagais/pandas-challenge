## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#total number of unique items
total_items = purchase_data["Item ID"].nunique()

#Average price 
average_price = round(purchase_data["Price"].mean(),2)

#Number of purchase
number_purchases =len(purchase_data["Purchase ID"])

#Total revenue 
total_revenue = purchase_data["Price"].sum()
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_df = purchase_data.groupby("Gender")

#Count for number of purchases of gender 
purchase_count = purchase_df["Purchase ID"].count()

#Average purchase price 
avg_price = round(purchase_df["Price"].mean(),2)

#Total purchase value 
total_price = round(purchase_df["Price"].sum(),2)

#Average total per person 
avg_total=[]
gender_count = [81,484,11]

for c in range(len(gender_count)):
    value = round((total_price[c]/gender_count[c]),2)
    value = "$" + str(value)
    avg_total.append(value)
```

```python
#formatting the cells 
avg_price = avg_price.map("${:,.2f}".format)
total_price = total_price.map("${:,.2f}".format)
```

```python
#Create a Purchasing Analysis table
purchase_df = pd.DataFrame({"Purchase Count": purchase_count,"Average Purchase Price": avg_price, "Total Purchase Value": total_price,
                           "Avg Total Purchase per Person": avg_total})
purchase_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#Creating bins 
bins = [0,9,14,19,24,29,34,39,50]
age_groups = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

#Calculating total count
age_df = purchase_data.loc[:,["Gender","SN","Age"]].drop_duplicates()
age_df

age_df["Age group"] = pd.cut(purchase_data["Age"],bins,labels=age_groups)
total_count = age_df.groupby("Age group")["Gender"].count()

#Calculating percentage 
percent = round((total_count/total_players)*100,2)
```

```python
#summary table
age_summary = pd.DataFrame({"Total Count": total_count, "Percentage of Players": percent})
age_summary
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Getting the data
purch_df = purchase_data.loc[:,["Gender","SN","Age","Item ID", "Price"]]
```

```python
#Bin the data in different age groups 
bins = [0,9,14,19,24,29,34,39,50]
age_groups = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"] 

#calculate purchase count 
purch_df["Age group"] = pd.cut(purchase_data["Age"],bins,labels=age_groups)
purch_group = purch_df.groupby("Age group")

purch_count = purch_group["Item ID"].count()
```

```python
#calculate average purchase price 
purch_price = round(purch_group["Price"].mean(),2)

#total purchase value 
purch_val = round(purch_group["Price"].sum(),2)

#average total purchase per person 
total_purch = round(purch_val/total_count,2)
```

```python
#formatting the dataframe
purch_price = purch_price.map("${:,.2f}".format)
purch_val = purch_val.map("${:,.2f}".format)
total_purch = total_purch.map("${:,.2f}".format)
```

```python
#create a summary table 
purch_summary = pd.DataFrame({"Purhcase count": purch_count,"Average Purchase Price": purch_price, 
                              "Total purchase value": purch_val, "Avg Total Purchase per Person": total_purch})
purch_summary
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#group by "SN"
top = purchase_data.groupby("SN")
top_count = top["Item ID"].count()

#average price per person
total_price = top["Price"].sum()
avg_price = round(total_price/top_count,2)

#total purchase value 
total_val = top["Price"].sum()
```

```python
#formatting the cells 
avg_price = avg_price.map("${:,.2f}".format)
total_val = total_val.map("${:,.2f}".format)
```

```python
#create a summary table
purch_summary = pd.DataFrame({"Purchase Count": top_count, "Average Purchase Price": avg_price, 
                              "Total Purchase Value": total_val})

purch_summary.sort_values("Total Purchase Value", axis = 0, ascending = False, 
                 inplace = True) 
purch_summary
```

## Most Popular Items

* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting

```python
#Grouping by Item ID and Item Name #NEED HELP# 
popitem_df = purchase_data.loc[:,["Item ID","Item Name","Price"]]
pop_grouped = popitem_df.groupby(["Item ID", "Item Name"])

#creating needed series
pop_sum = pop_grouped.sum()["Price"]
pop_average = pop_grouped.mean()["Price"]
pop_count = pop_grouped.count()["Price"]

#creating a new dataframe
pop_summary = pd.DataFrame({"Purchase Count": pop_count, "Item Price": pop_average, 
                            "Total Purchase Value": pop_sum})
pop_summary
```
